```python
# Importing python libraries needed for this project

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset into DataFrame

dataset1=pd.read_csv('disney_plus_titles.csv') # CSV file attached in folder
dataset1
```

```python
# Missing data

dataset1.nunique() # Total number of Movies and TV shows = 1450
```

```python
NaN = dataset1[dataset1.isna().any(axis=1)].shape # Total number of missing values
NaN
```

```python
(NaN[0]/dataset1.shape[0]) * 100 # Percenatge of missing values

# With a large proportion of missing data, removing NaN might give a more accurate analysis
```

```python
# Removing NaN values

dataset1['director'].replace(np.nan,'Unknown',inplace=True)
dataset1['country'].replace(np.nan,'Unknown',inplace=True)
dataset1['rating'].replace(np.nan,'Unknown',inplace=True)
dataset1['title'].replace(np.nan,'Unknown',inplace=True)
dataset1['type'].replace(np.nan,'Unknown',inplace=True)
dataset1['cast'].replace(np.nan,'Unknown',inplace=True)
dataset1['date_added'].replace(np.nan,'Unknown',inplace=True)
dataset1['release_year'].replace(np.nan,'Unknown',inplace=True)
dataset1['listed_in'].replace(np.nan,'Unknown',inplace=True)
dataset1['description'].replace(np.nan,'Unknown',inplace=True)
dataset1
```

```python
# Movies vs TV Shows
# 73% of Disney+ content is Movies and 27% is TV Shows

plt.pie(x=dataset1['type'].value_counts(),
        colors=['Green','Yellow'],
        labels=['Movies','TV Shows'],
        autopct='%1.0f%%',
        startangle=90,
        explode=[0,0.1],
        shadow=True)
plt.title('Movies vs TV Shows on Disney+')
plt.legend(loc='upper left')
plt.show()
```

```python
# Ratings Distribution
# "TV-PG" is the most popular rating for both Movies and TV Shows on Disney+

chart1 = dataset1.groupby(['type','rating']).agg({'show_id':'nunique'}).reset_index().sort_values('show_id',ascending=False)
plt.figure(figsize=(15,5))
chart1=sns.barplot(data=chart1, x='rating', y='show_id', hue='type')
plt.xticks(rotation=45,fontsize=7)
plt.yticks(fontsize=7)
plt.xlabel('Ratings',fontsize=10)
plt.ylabel(ylabel='Total shows',fontsize=10)
plt.legend(loc='upper right')
plt.title('Ratings contributions for Movies and TV Shows')

for p in chart1.patches:
    chart1.annotate('{}'.format(p.get_height()),(p.get_x(),p.get_height()+15),fontsize=7)
plt.show()
```

```python
# Top Genre
# 'Animation', 'Comedy', 'Family' is the most popular genre on Disney+

chart2 = dataset1.groupby(['listed_in']).agg({'show_id':'nunique'}).reset_index().sort_values('show_id',ascending=False).head(15)
plt.figure(figsize=(15,5))
chart2=sns.barplot(data=chart2,x='show_id', y='listed_in')
plt.ylabel('Genre',fontsize=10)
plt.xlabel(xlabel='Total Number of Movies and TV Shows',fontsize=10)
plt.title('Top Genre on Disney+')
```

```python
# Function for colllecting the Top 10 actors appearing in most Disney+ Movies and TV Shows.

def parse_column(column):
    column = column.str.split(',')

    series = list()
    for index, value in column.items():
        series.append(value)

    words = list()
    counts = dict()
    for obj in series:
        for word in obj:
            word = str(word).lstrip()
            words.append(word)

    for word in words:
        if word not in counts:
            counts[word] = 1
        else:
            counts[word] += 1
    counts_items = counts.items()
    data_list = list(counts_items)
    dataset2 = pd.DataFrame(data_list)
    dataset2 = dataset2.rename({0: column.name, 1: 'total'}, axis=1)
    dataset2 = dataset2.sort_values(by=['total'], ascending=False)
    dataset2 = dataset2.reset_index(drop=True)
    return dataset2
```

```python
# Jim Cummings has appeared in most Disney+ Movies and TV Shows
# 190 Movies and TV Shows where cast members are not specified. Need to update cast entries in database.

most = dataset1['cast'].notnull()
disney_castdf = dataset1['cast'][most]
disney_cast = parse_column(disney_castdf)
disney_cast[:10]
```
